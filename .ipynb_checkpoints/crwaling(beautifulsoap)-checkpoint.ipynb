# 웹 크롤링 

### 나의 경우 엑셀에서 추출한 제품명의 최저가를 웹에서 추출하기 위해 크롤링을 구현
따라서 아래 코드는 엑셀에서 원하는 컬럼을 가져온다. 엑셀에서 원하는 컬럼을 추출하는 예제는 아래 링크를 참고하기 바란다 <br>
https://github.com/park-moonkyu/oepnpyxl 

```python
# -*- coding:utf-8 -*-
import openpyxl

# 엑셀파일 열기
filename = "test.xlsx"
book = openpyxl.load_workbook(filename)

# 맨 앞의 시트 추출하기
sheet = book.worksheets[0]

# 시트의 각 행을 순서대로 추출하기
data = []
for row in sheet.rows:
    data.append([row[0].value, row[1].value,row[2].value])

del data[0]

for i in range(len(data)):
    print(data[i])

    
```

## soup 만들기

```python
# BeautifulSoup(태그, "해석라이브러리") > 뷰숲 공식문서에서 해석기에 대한 설명을 보자.
# 인자로 넣은 문서는 유니코드로 변환되며, html 개체는 유니코드 문자로 변환된다.
soup = BeautifulSoup(html, 'html.parser')
```

## findAll() & find()
findAll(name, attrs, recursive, text, limit, **kwargs) <br>
전체 태그를 훓어보게 된다.<br><br>
find(name, attrs, recursive, text, **kwargs) <br>
단 1개만 갖는 태그를 검색할때 사용한다. findAll보다 비용이 적게 발생하기 때문에 findAll('body' limit=1)보다 find('body')가 더 좋다

### 태그로 탐색
name: 태그, 정규표현식, 리스트, 함수, bool 같은 인자를 말한다.<br>
인자를 여러가지 조합해 해당하는 것을 추려낼 수 있다.<br><br>

attrs : css 클래스를 입력하면 해당 클래스를 갖는 데이터를 리턴한다. 하지만 이 방법보단 키워드 인자가 유용



```python
soup.find_all("a",class_="subFilter_tab__2cAf2")
```

###  다양한 방법으로 긁어오기
특정 태그를 사용하는 데이터를 추출하려면 태그인자를 사용한다. 태그는 문자열로 입력되어야 한다.<br>

### 정규표현식 인자로 넣어 b로 시작하는 모든 태그 찾기
soup.fild_all(re.complie('^b')) # b로 시작하는 모든 태그를 찾는다.


```python
# 리스트에 담긴 문자열이 해당하는 태그를 리스트에 담아 리턴하기
soup.find_all(['p','span'])
# [
# <p> 이건 p태그 </p>,
# <span> 이건 span1 </span>,
# <span> 이건 span2 </span>,
# ] 

# booL : T/F값을 인자로 줄수 있으며 True일때 모든 태그의 이름만 리턴한다. 태그가 포함하는 문자열은 출력하지않으며 tag.name만 출력
soup.find_all(True)
```

## 함수 
위 모든 인자들로 찾을 수 없으면 함수를 정의해서 사용한다.<br>


```python
# beautifulsoup 공식문서의 예제코드 퍼옴
from bs4 import NavigableString
def surrounded_by_strings(tag):
    return (isinstance(tag.next_element, NavigableString)
            and isinstance(tag.previous_element, NavigableString))

for tag in soup.find_all(surrounded_by_strings):
    print (tag.name)
```

```python
#id 값으로 찾기
soup.find_all(id='link')

# # 정규식으로 특정 문자가 들어간 값 찾기
# soup.find_all(href=re.compile("#"))

# #응용 두개 다 만족하는 값 찾기
soup.find_all(href=re.compile("#"), class_="style_inner__18zZX")
```
