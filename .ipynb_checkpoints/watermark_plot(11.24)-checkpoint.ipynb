```python
# watermark image

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
im_wm = plt.imread("./daewon.png")
plt.imshow(im_wm)
plt.show()
```

```python
fig, axes = plt.subplots(ncols=1, figsize=(4,4))

species_u = penguins["species"].unique()

for i ,s in enumerate(species_u):
    axes.scatter(penguins["bill_length_mm"].loc[penguins["species"]==s],
                   penguins["bill_depth_mm"].loc[penguins["species"]==s],
                      alpha=0.3
                   )
    
axes.legend(species_u, title="species") 
axes.set_xlabel("Bill Length (mm)")
axes.set_ylabel("Bill Depth (mm)")


figimage = fig.figimage(im_wm,resize=False, alpha=0.1,zorder=1)
fig.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axe = plt.subplots(figsize=(8,5))

sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", color="R", data=penguins, alpha=0.5, ax=axe, legend=False,zorder=2)
sns.kdeplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=penguins, alpha=0.7, ax=axe, legend=False,zorder=3)

figimage= fig.figimage(im_wm,resize=False,alpha=0.2,zorder=1)
```
