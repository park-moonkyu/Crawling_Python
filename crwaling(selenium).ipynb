## 셀리니움 

```python
from selenium import webdriver
import time


browser = webdriver.Chrome()
browser.get('https://auto.naver.com/bike/mainList.nhn')  

# 타이틀 가지고 오기
print(browser.title)

# 현재 url 출력
print(browser.current_url)

#  id일때는 # , class일때는 div. 을 사용한다. 그리고 마지막엔 요소를 a , span 등
browser_a= browser.find_element_by_css_selector("#container > div.spot_main > div.spot_aside > div.tit > a")
browser_a.click()

time.sleep(3)

# 바이크 제조사 1번째 페이크 진앱해서 리스트 추출
allBikeCompanyElement = browser.find_elements_by_css_selector(
    "#_vendor_select_layer > div > div.maker_group div.emblem_area > ul > li")
   
for item in allBikeCompanyElement:
    bikeComName = item.find_element_by_tag_name("span").text
    if (bikeComName != ''):
        print("바이크 회사명:" + bikeComName)
        ahref = item.find_element_by_tag_name("a").get_attribute("href")
        print('네이버 자동차 바이크제조사 홈 sub url:', ahref)
        imgUrl = item.find_element_by_tag_name("img").get_attribute("src")
        print('바이크 회사 엠블럼:', imgUrl)
        
time.sleep(3)

# 바이크 제조사 리스트의 다음 페이지 버튼을 찾아서 클릭하자.
nextBtn = browser.find_element_by_css_selector(
    "#_vendor_select_layer > div > div.maker_group > div.rolling_btn > button.next")

#바이크 제조사의 활성화 여부를 체크하는 코드이다.
isExistNextPage = nextBtn.is_enabled()

if (isExistNextPage == True):
    print("다음 페이지 존재함=======================================>")
    nextBtn.click()
    allBikeCompanyElement = browser.find_elements_by_css_selector(
        "#_vendor_select_layer > div > div.maker_group div.emblem_area > ul > li")
    for item in allBikeCompanyElement:
        bikeComName = item.find_element_by_tag_name("span").text
        if (bikeComName != ''):
            print("바이크 회사명:" + bikeComName)
            ahref = item.find_element_by_tag_name("a").get_attribute("href")
            print('네이버 자동차 바이크제조사 홈 sub url:', ahref)
            imgUrl = item.find_element_by_tag_name("img").get_attribute("src")
            print('바이크 회사 엠블럼:', imgUrl)
            
   
```
